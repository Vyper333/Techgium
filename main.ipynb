```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import folium
```

```python
df = pd.read_csv('systhetic_data_3.csv')
df['Date_Time'] = pd.to_datetime(df['Date_Time'])
```

```python
df.head()
```

```python
from folium import plugins
from folium.plugins import HeatMapWithTime
```

```python
map_hooray = folium.Map(location=[15.292169, 73.969181],zoom_start=13)

df['Latitude'] = df['Latitude'].astype(float)
df['Longitude'] = df['Longitude'].astype(float)

# heat_df = df[df['Date_Time'] == '09:00:00']

heat_df = df[['Latitude', 'Longitude']]
heat_df = heat_df.dropna(axis=0, subset=['Latitude', 'Longitude'])

time_index = []
for i in df['Date_Time'].unique():
    time_index.append(i)

# date_strings = [d.strftime('%d/%m/%Y, %H:%M:%S') for d in time_index]
```

```python
heat_data = [[row['Latitude'], row['Longitude']]for index, row in heat_df.iterrows()]

HeatMapWithTime(heat_data, auto_play=True,index=df['Date_Time'], max_opacity=0.7).add_to(map_hooray)

map_hooray
```
